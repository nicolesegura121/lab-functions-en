## 1. Write a function that returns the greater of two numbers

```python
def greater(a,b):
    if type(a) == int: 
        if type(b) == int: 
            if a < b : 
                return b
            if a > b:
                return a
            if a == b: 
                return f"Numbers must be different"
        else:
            return f"Input must be integer"
    else:
        return f"Input must be integer" 
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    result=1
    for i in range(1, n+1):
        result*=i
    return result
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

```python
#Recursive mode

def factorial2(n): 
    if n==0: 
        return 1
    elif n > 0 :
        result = n*factorial2(n-1)
        return result
```

```python
# This will test your function 
test_factorial(factorial2)
```

```python
def st_dev(list_sd):
    #single parameter
    x_avg = sum_all(list_sd)/len(list_sd)
    squared_x_var = []
    for x in list_sd: 
        squared_x_var.append((x-x_avg)**2)

    std = (sum_all(squared_x_var)/(len(list_sd)-1))**(1/2)
  
    return std

```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
import string
string.ascii_lowercase = set(string.ascii_lowercase)
type(string.ascii_lowercase)
```

```python
def pangram(sentence):
    #Importing an alphabet
    import string
    string.ascii_lowercase = set(string.ascii_lowercase)
    #Simplifying the sentence if there is any uppercase
    sentence = sentence.lower()
    #Defining an empty set to collect all letters present in the sentence
    letters = set()

    for letter in sentence: 

        if letter in string.ascii_lowercase: 

            letters.add(letter)

    return string.ascii_lowercase.issubset(letters)
```

```python
sentence2 = "The quick brown fox jumps over the lazy dog"
pangram(sentence2)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! 🤪`

```python
#I will generate a function that goes through all characters and stop until a comma is encountered
def sort_alpha(string):
    lst = []
    string_modified =""

    for character in string:
        if character!=",":
            string_modified+=character
        else:
            lst.append(string_modified.strip())
            string_modified =""
    lst.append(string_modified.strip())
    lst.sort()    
    return lst
```
